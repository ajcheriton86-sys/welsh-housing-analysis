```python
filtered = df[
    (df['Area'] =='Wales') &
    (df['Period'].str.contains('Annual')) &
    (df['Data description'] == 'All households') &
    (df['Outcomes'] == 'Total Outcomes')
    ]
filtered
```

```python
import matplotlib.pyplot as plt
filtered.sort_values('Period').plot(
    x='Period', 
    y='Data values',
    kind='line',
    figsize=(10, 6),
    title='Total homelessness outcomes in Wales, 2015-16 to 2024-25',
    legend=False
)
plt.xlabel('Reporting year')
plt.ylabel('Number of household cases handled')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

    
```
